```python
import pandas as pd
import matplotlib.pyplot as plt

MLBClosersData = {
"Name": ["C. Estevez", "A. Chapman", "R. Suarez", "K. Jansen", "E. Pagan", "E. Clase"],
"Inn. Pitched": [66.0, 61.1, 69.2, 59.0, 68.2, 47.1],
"ERA": [2.45, 1.17, 2.97, 2.59, 2.88, 3.23],
"Saves": [42, 32, 40, 29, 32, 24],
"League": ["AL", "AL", "NL", "AL", "NL", "AL"]
}
df= pd.DataFrame(MLBClosersData)
### Fun Fact: This was E. Clase's last season he was caught throwing games
print("MLBClosersData")
print(df)
n=3 #I picked 3 for example
print("MLBClosersData - Head (n=3)")
print(df.head(n))
print("MLBClosersData - Info")
print(df.info())
print("MLBClosersData - Describe")
print(df.describe())
```

Markdown 1: I noticed that all the commands worked, and I like how the describe data shows both mean and standard deviation of these closers' ERAs.

```python
print("Select 1 numerical column (ERA)")
print(df["ERA"])
print("Select 2 or more columns")
print(df[["Name", "Saves"]])
print("When Innings Pitched greater than 62")
print(df[df["Inn. Pitched"] > 62 ])
print("Categorical Filter when League is NL")
dt = df[df["League"] == "NL"]
print(dt)
print("Adding New Stats")
df["Blown Saves"] = [6, 2, 5, 1, 6, 5]
df["Saves per Game %"] = df["Saves"]/(df["Saves"]+df["Blown Saves"])
print(df)
print("Sort top 5 by Saves Per Game %")
dp= df.sort_values(by= "Saves per Game %", ascending= False)
print(dp.head(5))
```

Markdown 2: We see that when sorting the top 5 of S/G %, Kenley Jansen was most efficent

```python
print("Average Closer ERA by League in DataFrame")
print(df.groupby("League")["ERA"].mean())
print("Scatter Plotting ERA vs Inn. Pitched")
plt.figure()
plt.scatter(df["Inn. Pitched"], df["ERA"])
plt.xlabel("Inn. Pitched")
plt.ylabel("ERA")
plt.title("ERA vs Inn. Pitched")
plt.show()
print("Bar Graph of Saves")
plt.figure()
plt.bar(df["Name"], df["Saves"])
plt.xlabel("Name")
plt.ylabel("Saves")
plt.title("Saves per Person")
plt.show()
```
